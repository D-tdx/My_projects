```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())
```

Describing

```python
df.describe()
```

Gathering Information 

```python
df.info()
```

Checking is there any null values or not

```python
df.isnull().sum()
```

Delete unnamed coloumn

```python
df=df.drop("Unnamed: 0" ,axis = 1)
print(df.head())
```

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("5 - 10", "<10")
print(df.head())
```

Gender distribution

```python
plt.figure(figsize=(3,5))
ax=sns.countplot(data= df,x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

from the upper graph we can analyse that:
# The number of females are more than the number of males.

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

Parents' Eduction empact on students

```python
plt.figure(figsize=(8,5))
sns.heatmap(gb, annot = True)
plt.title("Relation between student's score and their parents' Education")
plt.show()
```

from the heatmap we can conclude that:
# There is certain impact on students's score, who have parents with good education.

```python
xc=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(xc)
```

```python
plt.figure(figsize=(8,3))
sns.heatmap(xc, annot = True)
plt.title("Relation between student's score and their parents' martial status")
plt.show()
```

from the heatmap we can conclude that:
# There is no or negligible certain impact on students's score due to their parents' martial; status.

Distribution of Ethnic Group

```python
print(df["EthnicGroup"].unique())
```

```python
GroupA = df.loc[(df["EthnicGroup"]== "group A")].count()
GroupB = df.loc[(df["EthnicGroup"]== "group B")].count()
GroupC = df.loc[(df["EthnicGroup"]== "group C")].count()
GroupD = df.loc[(df["EthnicGroup"]== "group D")].count()
GroupE = df.loc[(df["EthnicGroup"]== "group E")].count()
l=["GroupA","GroupB","GroupC","GroupD","GroupE"]
mlist=[GroupA["EthnicGroup"],GroupB["EthnicGroup"],GroupC["EthnicGroup"],GroupD["EthnicGroup"],GroupE["EthnicGroup"]]
plt.pie(mlist,labels=l, autopct="%1.2f%%")
print(mlist)
plt.title("Distribution of Ethnic Groups")
plt.show()
```
